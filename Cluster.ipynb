```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Cargar el archivo con las ubicaciones
file_path = 'C:/Users/enina/Desktop/Business Analytics/Segmentacion/data_ecu.txt'
data = pd.read_csv(file_path, delim_whitespace=True, header=0)

# Limpiar los datos: Reemplazar comas por puntos en latitudes y longitudes
data['Longitude'] = data['Longitude'].str.replace(',', '.')
data['Latitude'] = data['Latitude'].str.replace(',', '.')

# Convertir las columnas a números, manejando errores
data['Longitude'] = pd.to_numeric(data['Longitude'], errors='coerce')
data['Latitude'] = pd.to_numeric(data['Latitude'], errors='coerce')

# Eliminar filas con valores NaN (errores en la conversión)
data.dropna(inplace=True)

# Aplicar K-means para 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(data[['Longitude', 'Latitude']])

# Agregar las etiquetas de los clusters al DataFrame
data['Cluster'] = kmeans.labels_

# Visualización de los clusters
plt.figure(figsize=(8, 6))
plt.scatter(data['Longitude'], data['Latitude'], c=data['Cluster'], cmap='viridis')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.title('Clusters geográficos de ubicaciones en Ecuador')
plt.colorbar(label='Cluster')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Cargar el archivo con las ubicaciones
file_path = 'C:/Users/enina/Desktop/Business Analytics/Segmentacion/data_ecu.txt'
data = pd.read_csv(file_path, delim_whitespace=True, header=0)

# Verificar las columnas
print(data.columns)

# Limpiar los nombres de las columnas
data.columns = data.columns.str.strip()

# Limpiar los datos: Reemplazar comas por puntos en latitudes y longitudes
data['Longitude'] = data['Longitude'].str.replace(',', '.')
data['Latitude'] = data['Latitude'].str.replace(',', '.')

# Convertir las columnas a números, manejando errores
data['Longitude'] = pd.to_numeric(data['Longitude'], errors='coerce')
data['Latitude'] = pd.to_numeric(data['Latitude'], errors='coerce')

# Eliminar filas con valores NaN (errores en la conversión)
data.dropna(inplace=True)

# Aplicar K-means para 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(data[['Longitude', 'Latitude']])

# Agregar las etiquetas de los clusters al DataFrame
data['Cluster'] = kmeans.labels_

# Crear un archivo de salida con la estructura "location_id cluster_label"
output_file = 'C:/Users/enina/Desktop/Business Analytics/Segmentacion/data_ecu_clusters.txt'
with open(output_file, 'w') as f:
    for index, row in data.iterrows():
        f.write(f"{index} {row['Cluster']}\n")

print("Archivo data_ecu_clusters.txt generado correctamente.")
```
